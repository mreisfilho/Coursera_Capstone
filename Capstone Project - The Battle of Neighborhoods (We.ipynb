### 1. A description of the problem and a discussion of the background
    
    - The problem here is the difficulty in opening branches around the world. It is widely recognized that opening international braches can be very positive for Brand expansion around the world and consequently leverage profit.
    - However. Due to dissimilarity of country's cultures and also geolocation size, radius, etc., It is not unusual a very successfull Brand don't succeed in a new environment.
    - The approach in this notebook will be to find similar neighbourhoods using foursquare location, in order to save money on field research. And to give a simple and fast solution for Business Executives to consider if they want to expand their Brand overseas.
    - The main goal is to categorize the neighbourhoods in Sao Paulo, Brazil. In order to make the decision of where to open a new business an easy decision, or better, a logical decision. Using Data, of course
